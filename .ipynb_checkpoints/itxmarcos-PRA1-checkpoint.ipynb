Se deben eliminar esas filas con valores perdidos.
# Preprocesado y gestión de características
## Paso 3: limpiar los datos
Ahora que hemos identificado varios errores en el conjunto de datos, tenemos que corregirlos antes de continuar con el análisis. Revisemos los problemas uno por uno.

Puede que a los investigadores de campo se les olvidara añadir Iris- antes de sus entradas Iris-versicolor. La otra clase extraña, Iris-setossa, es un error tipográfico.

```python
iris_data.loc[iris_data['class'] == 'versicolor', 'class'] = 'Iris-versicolor'
iris_data.loc[iris_data['class'] == 'Iris-setossa', 'class'] = 'Iris-setosa'

iris_data['class'].unique()
```

En el caso de la entrada anómala de Iris-setosa, es poco probable que Iris-setosa tenga un ancho de sépalo inferior a 2,5 cm. Seguramente este dato se introdujera por error, y es mejor eliminar la entrada que intentar averiguar realmente lo que pasó.

```python
# Dropeamos las filas Iris-setosa con sepal_width < 2.5 cm
iris_data = iris_data.loc[(iris_data['class'] != 'Iris-setosa') | (iris_data['sepal_width_cm'] >= 2.5)]
iris_data.loc[iris_data['class'] == 'Iris-setosa', 'sepal_width_cm'].hist()
;
```

Ahora todas nuestras filas de Iris-setosa tienen un ancho de sépalo superior a 2,5.

El siguiente problema de datos a tratar son las varias longitudes de los sépalos cercanas a cero para las filas de Iris-versicolor. 

```python
iris_data.loc[(iris_data['class'] == 'Iris-versicolor') &
              (iris_data['sepal_length_cm'] < 1.0)]
```

Todas estas entradas de sepal_length_cm cercanas a cero parecen estar desviadas en dos órdenes de magnitud, como si se hubieran registrado en metros en lugar de en centímetros.

```python
iris_data.loc[(iris_data['class'] == 'Iris-versicolor') &
              (iris_data['sepal_length_cm'] < 1.0),
              'sepal_length_cm'] *= 100.0

iris_data.loc[iris_data['class'] == 'Iris-versicolor', 'sepal_length_cm'].hist()
;
```

Ahora echemos un vistazo a las filas con missing values:

```python
iris_data.loc[(iris_data['sepal_length_cm'].isnull()) |
              (iris_data['sepal_width_cm'].isnull()) |
              (iris_data['petal_length_cm'].isnull()) |
              (iris_data['petal_width_cm'].isnull())]
```

Cuando no se puede permitir perder datos (droppear filas) o los datos que faltan son sistemáticos (todos los valores que faltan están en la misma columna para el mismo tipo de Iris), este error podría sesgar nuestro análisis.

Se puede recurrir a la imputación de la media: Si sabemos que los valores de una medida están dentro de un rango determinado, podemos rellenar los valores vacíos con la media de esa medida.

```python
iris_data.loc[iris_data['class'] == 'Iris-setosa', 'petal_width_cm'].hist()
;
```

La mayoría de las anchuras de los pétalos de Iris-setosa se sitúan en el rango de 0,2-0,3, por lo que vamos a completar estas entradas con la anchura media de los pétalos medida.

```python
#iris_data.dropna(inplace=True)

average_petal_width = iris_data.loc[iris_data['class'] == 'Iris-setosa', 'petal_width_cm'].mean()

iris_data.loc[(iris_data['class'] == 'Iris-setosa') &
              (iris_data['petal_width_cm'].isnull()),
              'petal_width_cm'] = average_petal_width

iris_data.loc[(iris_data['class'] == 'Iris-setosa') &
              (iris_data['petal_width_cm'] == average_petal_width)]
```

```python
iris_data.loc[(iris_data['sepal_length_cm'].isnull()) |
              (iris_data['sepal_width_cm'].isnull()) |
              (iris_data['petal_length_cm'].isnull()) |
              (iris_data['petal_width_cm'].isnull())]
```

Ahora hemos recuperado esas filas y ya no tenemos missing values en el conjunto de datos.

## Paso 4: Análisis exploratorio
Trataremos de responder a preguntas como ¿Cómo se distribuyen mis datos? ¿Existen correlaciones en mis datos? ¿Existen motivos que expliquen estas correlaciones? Nuestros datos se distribuyen normalmente en su mayor parte, lo cual es una gran noticia si planeamos utilizar cualquier método de modelado que asuma que los datos se distribuyen normalmente.

En la gráfica anterior, observamos que las mediciones de los pétalos quizá tengan algo que ver con los diferentes tipos de Iris.
- La distribución de las medidas de los pétalos se debe a las diferentes especies. Esto facilita enormemente la tarea de clasificación, ya que significa que las medidas de los pétalos permitirán distinguir fácilmente entre Iris-setosa y los demás tipos de Iris. Distinguir entre Iris-versicolor e Iris-virginica resultará más difícil, ya que sus medidas se solapan.
- También existen correlaciones entre la longitud y la anchura de los pétalos, así como entre la longitud y la anchura de los sépalos. Los biólogos de campo aseguran que esto es de esperar: Los pétalos de las flores más largos también tienden a ser más anchos, y lo mismo ocurre con los sépalos.

Podemos hacer gráficos de violín de los datos para comparar las distribuciones de las medidas de las clases. Estos, en comparación con los diagramas de caja, además escalan la caja según la densidad de los datos.

```python
plt.figure(figsize=(10, 10))

for column_index, column in enumerate(iris_data_clean.columns):
    if column == 'class':
        continue
    plt.subplot(2, 2, column_index + 1)
    sb.violinplot(x='class', y=column, data=iris_data_clean)
```
