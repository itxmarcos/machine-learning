# Preprocesado y gestin de caractersticas
## Paso 3: limpiar los datos
Ahora que hemos identificado varios errores en el conjunto de datos, tenemos que corregirlos antes de continuar con el anlisis. Revisemos los problemas uno por uno.

Puede que a los investigadores de campo se les olvidara aadir Iris- antes de sus entradas Iris-versicolor. La otra clase extraa, Iris-setossa, es un error tipogrfico.

```python
iris_data.loc[iris_data['class'] == 'versicolor', 'class'] = 'Iris-versicolor'
iris_data.loc[iris_data['class'] == 'Iris-setossa', 'class'] = 'Iris-setosa'

iris_data['class'].unique()
```

En el caso de la entrada anmala de Iris-setosa, es poco probable que Iris-setosa tenga un ancho de spalo inferior a 2,5 cm. Seguramente este dato se introdujera por error, y es mejor eliminar la entrada que intentar averiguar realmente lo que pas.

```python
# Dropeamos las filas Iris-setosa con sepal_width < 2.5 cm
iris_data = iris_data.loc[(iris_data['class'] != 'Iris-setosa') | (iris_data['sepal_width_cm'] >= 2.5)]
iris_data.loc[iris_data['class'] == 'Iris-setosa', 'sepal_width_cm'].hist()
;
```

Ahora todas nuestras filas de Iris-setosa tienen un ancho de spalo superior a 2,5.

El siguiente problema de datos a tratar son las varias longitudes de los spalos cercanas a cero para las filas de Iris-versicolor. 

```python
iris_data.loc[(iris_data['class'] == 'Iris-versicolor') &
              (iris_data['sepal_length_cm'] < 1.0)]
```

Todas estas entradas de sepal_length_cm cercanas a cero parecen estar desviadas en dos rdenes de magnitud, como si se hubieran registrado en metros en lugar de en centmetros.

```python
iris_data.loc[(iris_data['class'] == 'Iris-versicolor') &
              (iris_data['sepal_length_cm'] < 1.0),
              'sepal_length_cm'] *= 100.0

iris_data.loc[iris_data['class'] == 'Iris-versicolor', 'sepal_length_cm'].hist()
;
```

Ahora echemos un vistazo a las filas con missing values:

```python
iris_data.loc[(iris_data['sepal_length_cm'].isnull()) |
              (iris_data['sepal_width_cm'].isnull()) |
              (iris_data['petal_length_cm'].isnull()) |
              (iris_data['petal_width_cm'].isnull())]
```

Cuando no se puede permitir perder datos (droppear filas) o los datos que faltan son sistemticos (todos los valores que faltan estn en la misma columna para el mismo tipo de Iris), este error podra sesgar nuestro anlisis.

Se puede recurrir a la imputacin de la media: Si sabemos que los valores de una medida estn dentro de un rango determinado, podemos rellenar los valores vacos con la media de esa medida.

```python
iris_data.loc[iris_data['class'] == 'Iris-setosa', 'petal_width_cm'].hist()
;
```

La mayora de las anchuras de los ptalos de Iris-setosa se sitan en el rango de 0,2-0,3, por lo que vamos a completar estas entradas con la anchura media de los ptalos medida.

```python
#iris_data.dropna(inplace=True)

average_petal_width = iris_data.loc[iris_data['class'] == 'Iris-setosa', 'petal_width_cm'].mean()

iris_data.loc[(iris_data['class'] == 'Iris-setosa') &
              (iris_data['petal_width_cm'].isnull()),
              'petal_width_cm'] = average_petal_width

iris_data.loc[(iris_data['class'] == 'Iris-setosa') &
              (iris_data['petal_width_cm'] == average_petal_width)]
```

```python
iris_data.loc[(iris_data['sepal_length_cm'].isnull()) |
              (iris_data['sepal_width_cm'].isnull()) |
              (iris_data['petal_length_cm'].isnull()) |
              (iris_data['petal_width_cm'].isnull())]
```

Ahora hemos recuperado esas filas y ya no tenemos missing values en el conjunto de datos.

Trataremos de responder a preguntas como Cmo se distribuyen mis datos? Existen correlaciones en mis datos? Existen motivos que expliquen estas correlaciones? Nuestros datos se distribuyen normalmente en su mayor parte, lo cual es una gran noticia si planeamos utilizar cualquier mtodo de modelado que asuma que los datos se distribuyen normalmente.

En la grfica anterior, observamos que las mediciones de los ptalos quiz tengan algo que ver con los diferentes tipos de Iris.
- La distribucin de las medidas de los ptalos se debe a las diferentes especies. Esto facilita enormemente la tarea de clasificacin, ya que significa que las medidas de los ptalos permitirn distinguir fcilmente entre Iris-setosa y los dems tipos de Iris. Distinguir entre Iris-versicolor e Iris-virginica resultar ms difcil, ya que sus medidas se solapan.
- Tambin existen correlaciones entre la longitud y la anchura de los ptalos, as como entre la longitud y la anchura de los spalos. Los bilogos de campo aseguran que esto es de esperar: Los ptalos de las flores ms largos tambin tienden a ser ms anchos, y lo mismo ocurre con los spalos.

Podemos hacer grficos de violn de los datos para comparar las distribuciones de las medidas de las clases. Estos, en comparacin con los diagramas de caja, adems escalan la caja segn la densidad de los datos.

```python
plt.figure(figsize=(10, 10))

for column_index, column in enumerate(iris_data_clean.columns):
    if column == 'class':
        continue
    plt.subplot(2, 2, column_index + 1)
    sb.violinplot(x='class', y=column, data=iris_data_clean)
```
