```python
# Imports
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
boston=load_boston()
print(boston.DESCR)
```

Formula para mininizar el error cuadratico medio (MCO): $\beta = (X^{T}X)^{-1}X^{T}Y$

```python
X=boston.data[:, 5]
Y=boston.target
plt.scatter(X,Y, alpha=0.3)

X.shape #Te da la forma de la matriz
X.T.shape #Te da la forma de su traspuesta

#AÃ±adimos columna de 1s para termino independiente
X=np.array([np.ones(506), X]).T
B=np.linalg.inv(X.T @ X)@ X.T @ Y
plt.plot([4,9], [B[0]+B[1]*4, B[0]+B[1]*9], c='red')
plt.show()
```
